After that we looked at the images in our dataset and decided that the color was unimportant/extra information. So we converted the images to grayscale before feeding to model. Of course, we also did normalization while doing this so that the input values have a similar distribution. Before I show the code, I want to mention a small detail. When converting to grayscale, the pictures go down to 1 dimension from 3 dimensions, but inceptionV3 wants 3 dimensions. Because of that, after converting to grayscale we converted back to RGB, but of course the colors are not coming back. You can find the code that I have used in [[4](#four)]
Here are the codes:(from "InceptionV3.py")                                                                                 
Note: We used rescale parameter for normalization

```python
from keras.applications.inception_v3 import InceptionV3
from keras.optimizers import RMSprop
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
from keras.callbacks import ModelCheckpoint, EarlyStopping
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Model

IMG_SIZE = 330


def to_grayscale_then_rgb(image):
    image = tf.image.rgb_to_grayscale(image)
    image = tf.image.grayscale_to_rgb(image)
    return image


tr_data = ImageDataGenerator(rescale=1.0/255.0, fill_mode="nearest", preprocessing_function=to_grayscale_then_rgb,
                                     validation_split=0.2)
```

We are now ready to get the data to train. But how to do it?
                                                                                                                              
The function of the terraflow library, which we see in the code below, is a function that reads the data from the folders and uses the folder names as the label. We must write the path of the folder containing our folders in the parameter named "directory". Subset parameter is about which part of the split you want.                                                                                                                                                                                     
The most important part of this function is the batch_size parameter. Why?  
                                                                                                                              
If you put all the data in an array instead of batches, you will most likely run into hardware problems/limits while training or even before starting the training because the data is very large. We encountered this problem. 
In order to make k-fold, we need to insert the images into an array for a short time in the code we wrote, but when we use all the train data we have, around 20GB of ram is required for this process.                                                   
Note: Class mode is binary but this mod works in more than 2 categories
Note: This and the last one are from "InceptionV3.py"

```python
# you may need to change addresses
data = tr_data.flow_from_directory(directory="train", subset="training",
                                   seed=123, shuffle=True, target_size=(IMG_SIZE, IMG_SIZE),
                                   batch_size=100, class_mode="binary")
validation = tr_data.flow_from_directory(directory="train", subset="validation",
                                         seed=123, shuffle=True, target_size=(IMG_SIZE, IMG_SIZE),
                                         batch_size=100, class_mode="binary")
```

Before I get into k-fold, I want to talk about image size. If the image size we provide is larger than our data's image size, it uses the function we wrote in the parameter called fill_mode, one of the parameters of the ImageDataGenerator, to fill in the blanks. But our image size is 330, so it is smaller than our data's image size. We couldn't find documentation on how it is  shrinking the image in this case

We got help for k-fold from a post in stackoverflow (which you can find in [[5](#five)]).

## Modeling<a class="anchor" id="modeling"></a>

I mentioned that we are doing transfer learning using InceptionV3. Well, firstly, what is InceptionV3?                          
InceptionV3 is the third edition of Google's Inception Convolutional Neural Network. It is often used to aid image analysis.
The design principle is to provide a deeper networks with few parameters compared to others. For example, it has a parameter close to one-third of AlexNet


Our main reason for choosing it was that we knew from the study we found that this model had good results for our dataset.

Now that I explained our model, it's time to talk about transfer learning. Interestingly, the name of this technique fits perfectly. We can say that we borrow what the trained models have learned. 
That is, we add a few layers to the model we are using  for transfer learning according to the needs of us and give our data directly to the model. We make all layers untrainable except the layers we added ourselves so that we don't train them again.
This way we avoid the cost of training a lot of layers and frankly, since companies like Google have much larger data etc. than us, it is highly likely that if we want to train the model from zero, our training will give worse results compared to them. 
The code below is from "InceptionV3.py"

```python
pre_trained_model = InceptionV3(input_shape=(IMG_SIZE, IMG_SIZE, 3), include_top=False, weights="imagenet")
# [:289]: [:209]: 299 Some of the trials
for layer in pre_trained_model.layers[:289]:
    layer.trainable = False
layers = tf.keras.layers
```

We don't add top layer because InceptionV3's top layer is for output but we want to write output layer ourselves

I said all layers but here we make 289 layers untrainable, why?                                                               
Because we are doing fine tuning. Before I get into what fine tuning is, I have something to say.

According to some definitions, transfer learning is the use of a pre-trained model for another job without any change.
In this case, the layers we added for the output above are fine tuning.

So, fine tuning is:                                                                                                    
1- Changing the output of the pre-trained model                                                                                 
2- Retraining some pre-trained layers                                                                                           

```python
x = pre_trained_model.output
# Transfer learning is completed
x = layers.AveragePooling2D(pool_size=(8, 8), strides=1)(x)
x = layers.Flatten()(x)
x = layers.Dense(1024, activation="relu")(x)
x = layers.Dense(8, activation="softmax")(x)

model = Model(inputs=pre_trained_model.input, outputs=x)
```

Here we see how we change the output. 2048 feature maps that have size of 8x8 are coming from InceptionV3. we convert it to array of 2048 features using average pooling and then flatten layer. These features feed into 1,024 fully connected node layers and 8 fully connected node layers, respectively. We use the latest with 8 nodes because we have 8 categories

Average pooling is a pooling operation that calculates the average value for given size of feature maps one by one

```python
from IPython.display import Image
Image(url="https://pbs.twimg.com/media/FlEXvhBXkAAhO7V?format=png&name=large")
```

For example, here I show how a 4x4 feature map is pooled on average with a pool size of 2x2 and stride equal to 2. Stride is the number of pixels shifts for new average pooling calculation

Flatting layers are converting the data into a 1D array. On the other hand, the dense layers are simple layer of neurons in which each neuron receives input from all the neurons of previous layer, thus called as dense [[6](#six)]. For activation methods, you can look at the official documentation of Keras [[7](#seven)]. The Softmax is used for converts of values to a probability distribution.

```python
model.compile(optimizer=RMSprop(learning_rate=0.001), loss="sparse_categorical_crossentropy", metrics=["accuracy"])
checkpoint = ModelCheckpoint("CheckPoint/checkpoint", monitor='val_accuracy', verbose=1,
                             save_best_only=True, save_weights_only=True, mode='auto', save_freq='epoch')
early = EarlyStopping(monitor='val_accuracy', min_delta=0, patience=5, verbose=1, mode='auto')
# epochs = 100
epochs = 1
history = model.fit(data, validation_data=validation, epochs=epochs, verbose=1, callbacks=[checkpoint, early])
```

These are the final stages. We are compiling the model. We used RMSprop for optimizer and sparse_categorical_crossentropy for loss function. sparse_categorical_crossentropy should be used when we have mutually exclusive classes. If our target is integers like category indexes, we should use this.                                                                    
Root Mean Squared Propagation, or RMSProp, is an extension of gradient descent [[8](#eight)]. Learning rate is gradient descent's step size.                                                                                                
Since the training takes a long time, there can be a great loss in case of a mistake. To avoid this, we use Checkpoint. Checkpoint is used to save the model at some interval.                                                           
EarlyStopping, on the other hand, ends the training if a gain is not seen at the value we set as a result of the epochs.
When doing this modeling we also get help from [[9](#nine)] and [[10](#ten)]
                                                                                                                           
Note: An epoch means training the model with all the training data. If your epoch is set to 20, it means you want to train the Model 20 times with the same training data.
